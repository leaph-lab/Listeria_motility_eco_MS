```python
# Load data
df_distance = pd.read_excel(data_path + "T8_T12_clade_distance.xlsx")

# Set categorical ordering
df_distance['group'] = pd.Categorical(df_distance['group'], categories=df_distance['group'].unique(), ordered=True)
df_distance['group'] = df_distance['group'].cat.rename_categories({
    "T8": "L. welshimeri",
    "T12": "L. booriae"
})

df_distance['class'] = pd.Categorical(df_distance['class'], categories=df_distance['class'].unique(), ordered=True)
df_distance['color'] = pd.Categorical(df_distance['color'], categories=df_distance['color'].unique(), ordered=True)

palette = [
    "#EF5350", "#90A4AE", "#9575CD", "#80DEEA", '#4DB6AC', '#F57C00', 
    '#FFEB3B', '#9CCC65', '#BCAAA4', '#EC7063', '#E59866', '#F1C40F', 
    '#52BE80', '#73C6B6', '#7FB3D5', '#85C1E9', '#AF7AC5', '#ABB2B9'
]

plt.figure(figsize=(8, 6))

ax = sns.boxplot(
    data=df_distance,
    x='group',
    y='value',
    hue='color',
    palette=palette,
    dodge=True,
    width=2
)

# Calculate means
means = df_distance.groupby(['group', 'color'])['value'].mean().reset_index()

# Get unique groups and colors
groups = df_distance['group'].cat.categories
colors = df_distance['color'].cat.categories

# Number of colors (hue categories)
n_colors = len(colors)
width = 2  # boxplot width

# For each group and color, calculate x position and plot mean point
for i, group in enumerate(groups):
    for j, color in enumerate(colors):
        mean_val = means.loc[
            (means['group'] == group) & (means['color'] == color), 'value'
        ]
        if not mean_val.empty:
            # Calculate position of this box on x-axis:
            # Boxplot centers are at i (0-based for groups)
            # Hue boxes are spread inside width around center
            # Formula from seaborn docs: center  width/2 * (2*j/(n_colors-1) - 1)
            if n_colors > 1:
                offset = width / n_colors * j - width / 2 + width / (2 * n_colors)
            else:
                offset = 0

            x_pos = i + offset
            ax.scatter(
                x_pos,
                mean_val.values[0],
                color='white',
                edgecolor='black',
                s=50,
                zorder=10,
                linewidth=1,
                marker='o'
            )

ax.set_xlabel("")
ax.set_ylabel("Distance (km)", fontsize=16)
ax.set_title("Isolation distance of each clade", fontsize=18)
ax.tick_params(axis='x', labelsize=14)
ax.tick_params(axis='y', labelsize=12)
ax.get_legend().set_visible(False)

#plt.legend(title=None, bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=14, ncol=3)
plt.tight_layout()
plt.savefig(out_path_supp + 's1a_boo_wels_distance.pdf', format='pdf', dpi=1200)
plt.show()
```

```python
import os
import matplotlib.pyplot as plt
import seaborn as sns
from statannotations.Annotator import Annotator


def plot_2_box_with_mw_and_swarm(
    data,
    x,
    y,
    box_pairs=None,
    title="",
    figsize=(4, 6),
    colors=['#AF7AC5', '#5DADE2'],
    boxplot_kwargs=None,
    stat_test='Mann-Whitney',
    swarm_size=5,
    out_path="",
    out_filename=None,
    y_label=None,
    ax=None,
    enforce_order=("L. welshimeri", "L. booriae")
):
    """
    Box + swarm plot with Mann-Whitney + statannotations.
    Uses exact p-values (original behavior).
    """

    # Default boxplot styling
    if boxplot_kwargs is None:
        boxplot_kwargs = {
            'boxprops': {'edgecolor': 'k', 'linewidth': 2.5, 'facecolor': 'white'},
            'medianprops': {'color': 'k', 'linewidth': 2.5},
            'whiskerprops': {'color': 'k', 'linewidth': 2.5},
            'capprops': {'color': 'k', 'linewidth': 2.5},
            'width': 0.5
        }

    # Create axis if missing
    created_fig = False
    if ax is None:
        fig, ax = plt.subplots(figsize=figsize)
        created_fig = True

    # Clean data
    data = data.dropna(subset=[y])
    if data.empty:
        print(f"No data to plot for {title}.")
        return ax

    # Enforce species order
    present_groups = [g for g in enforce_order if g in data[x].unique()]
    if len(present_groups) < 2:
        present_groups = sorted(data[x].unique().tolist())

    # Color mapping
    palette = dict(zip(present_groups, colors[:len(present_groups)]))

    # Boxplot
    sns.boxplot(
        data=data,
        x=x,
        y=y,
        order=present_groups,
        palette=palette,
        ax=ax,
        showfliers=False,
        **boxplot_kwargs
    )

    # Swarmplot
    sns.swarmplot(
        data=data,
        x=x,
        y=y,
        order=present_groups,
        palette=palette,
        ax=ax,
        size=swarm_size,
        edgecolor='k',
        linewidth=0.5
    )

    # Default comparison
    if box_pairs is None and len(present_groups) >= 2:
        box_pairs = [(present_groups[0], present_groups[1])]

    # Statistical annotation
    annotator = Annotator(
        ax, box_pairs,
        data=data, x=x, y=y,
        order=present_groups
    )
    annotator.configure(
        test=stat_test,
        text_format='full',
        comparisons_correction=None,
        loc='inside',
        verbose=2,
        fontsize=12,
        pvalue_format_string="{:.1e}"
    )

    try:
        annotator.apply_and_annotate()
    except Exception as e:
        print(f"Annotation error: {e}")

    # Labels
    ax.set_title(title, fontsize=16)
    ax.set_xlabel("")
    ax.set_ylabel(y_label if y_label else y, fontsize=14)
    ax.tick_params(axis='x', labelsize=12)

    # Spines
    ax.grid(False)
    for spine in ax.spines.values():
        spine.set_edgecolor('black')
        spine.set_linewidth(0.7)

    # Save + show
    if created_fig:
        plt.tight_layout()

        if out_path:
            if out_filename is None:
                safe_title = title.replace("/", "_").replace("\\", "_")
                out_filename = f"{safe_title}.pdf"

            plt.savefig(
                os.path.join(out_path, out_filename),
                format="pdf",
                dpi=600
            )

        plt.show()

    return ax
```

```python
# Group by phylogroup and calculate the mean of 'value'
mean_of_phylogroup = df_distance.groupby('color')['value'].mean().reset_index()

# Optionally rename columns for clarity
mean_of_phylogroup.columns = ['color', 'mean_value']
mean_of_phylogroup = mean_of_phylogroup.rename(columns={'mean_value': 'Mean isolation distance of each phylogroup'})
mean_of_phylogroup[['species', 'clade']] = mean_of_phylogroup['color'].str.split('_', expand=True)
mean_of_phylogroup['species'] = mean_of_phylogroup['species'].replace({
    'T8': 'L. welshimeri',
    'T12': 'L. booriae'
})

# Plot using the subsampled data
plot_2_box_with_mw_and_swarm(
    data=mean_of_phylogroup,
    x="species",
    y="Mean isolation distance of each phylogroup",
    y_label="Distance (km)",
    box_pairs=[("L. booriae", "L. welshimeri")],
    title="Mean isolation distances",
    swarm_size=8,
    out_path=out_path_supp,
    out_filename='s1b_distance_compare.pdf'
)
```

```python
def assign_core_accessory(df, output_path, output_filename):
    """
    Process the input DataFrame, categorize rows based on the presence of '1' and '0', 
    and save the resulting DataFrame to a CSV file.

    Parameters:
    df (pd.DataFrame): The input DataFrame to be processed.
    output_path (str): The directory path where the CSV file will be saved.
    output_filename (str): The name of the CSV file.

    Returns:
    None
    """
    # Convert all values to strings and strip spaces
    df = df.applymap(lambda x: str(x).strip())

    # Remove rows where all cells (excluding the index) contain '0'
    df = df.loc[~(df == '0').all(axis=1)]

    # Categorize rows where all values are not '0' as 'Core'
    df.loc[~(df == '0').any(axis=1), 'cat'] = 'Core'

    # Assign 'Accessory' to remaining rows
    df['cat'].fillna('Accessory', inplace=True)

    # Export to CSV
    output_file = output_path + output_filename
    df.to_csv(output_file, index=False)
```

```python
# Function to split rows

def split_rows(df, column):
    # Split the 'COG Functional Category' column and create a new DataFrame with exploded rows
    df_expanded = df[column].apply(lambda x: pd.Series(list(x))).stack().reset_index(level=1, drop=True).to_frame(column)
    
    # Merge the expanded 'COG Functional Category' with the rest of the DataFrame
    df_expanded = df_expanded.join(df.drop(columns=[column]), how='left')
    
    return df_expanded
```

```python
# Function for FDR correction
import statsmodels.stats.multitest as smt

def fdr(p_value):
    p_v = []
    for c_key in p_value:
        p_v.append(p_value[c_key][0])
    array = np.array(p_v)
    rej, pvalscorr, _, _ = smt.multipletests(array, method = 'fdr_bh')
    p_corr = pvalscorr.tolist()
    n_p = 0
    fdr = {}
    for c_key in p_value:
        fdr[c_key] = p_corr[n_p]
        n_p = n_p + 1
    return fdr
```

```python
# Function for FDR correction - filter and keep only the significant FDR-adjusted P-vals

def fdr_sig(p_value):
    p_v = []
    for c_key in p_value:
        p_v.append(p_value[c_key][0])
    array = np.array(p_v)
    rej, pvalscorr, _, _ = smt.multipletests(array, method = 'fdr_bh')
    p_corr = pvalscorr.tolist()
    n_p = 0
    sig_fdr = {}
    for c_key in p_value:
        if p_corr[n_p] < 0.05:
            sig_fdr[c_key] = p_corr[n_p]
        n_p = n_p + 1
    return sig_fdr
```

```python
# Perform MW test
pval_mw = defaultdict(list)

for columnName, columnData in df_cog_prop_t.items():
    L1 = columnData[columnData.index.isin(boo)]
    L2 = columnData[columnData.index.isin(wels)]
    
    if len(L1) == 0 or len(L2) == 0:
        print(f"Skipping {columnName} due to empty group (L1: {len(L1)}, L2: {len(L2)})")
        continue

    # MW test between L1 and L2
    u_stat, pval = mannwhitneyu(L1, L2, alternative='two-sided')
    
    # Store the p-value in the dictionary
    pval_mw[columnName].append(pval)

fdr_cog = fdr(pval_mw)
sig_cog = fdr_sig(pval_mw)
len(sig_cog)
```

```python
def plot_volcano_with_stats(data, x, y, sig_data, title="", figsize=(4, 4), top_n=5, out_path="", xlim=None, ylim=None, save_as=None, point_size=20):
    plt.figure(figsize=figsize)
    ax = plt.subplot(111)

    # Non-significant points
    nonsig_data = data[~data.index.isin(sig_data.keys())]
    ax.scatter(nonsig_data[x], nonsig_data[y], color="#CFCFCF", alpha=0.5, edgecolors='k', linewidth=0.5, s=point_size)

    # Significant points
    sig_data_df = data[data.index.isin(sig_data.keys())]
    ax.scatter(sig_data_df[sig_data_df[x] > 0][x], sig_data_df[sig_data_df[x] > 0][y], color="#5DADE2", edgecolors='k', linewidth=0.5, s=point_size)
    ax.scatter(sig_data_df[sig_data_df[x] < 0][x], sig_data_df[sig_data_df[x] < 0][y], color="#AF7AC5", edgecolors='k', linewidth=0.5, s=point_size)

    # Annotate top N significant points
    top_sig_points = sig_data_df.nlargest(top_n, y)
    for idx, row in top_sig_points.iterrows():
        ax.annotate(idx, (row[x], row[y]), textcoords="offset points", xytext=(7, 0), ha='left')

    # Labels and title
    ax.set_xlabel("log2 (fold change)", fontsize=14)
    ax.set_ylabel("-log10 (adjusted P)", fontsize=14)
    ax.set_title(title, fontsize=16)

    # Customize tick label size
    ax.tick_params(axis='both', labelsize=12)

    # Add reference lines
    ax.axhline(y=-np.log10(0.05), color='#8C8C8C', linestyle='--')
    ax.axvline(x=0, color='#8C8C8C', linestyle='--')

    # Apply custom xlim if provided
    if xlim:
        ax.set_xlim(xlim)

    # Apply custom ylim if provided
    if ylim:
        ax.set_ylim(ylim)

    plt.tight_layout()

    if out_path:
        save_name = save_as if save_as else f'{title}.pdf'
        plt.savefig(out_path + save_name, format='pdf', dpi=300)

    plt.show()
```

```python
# Volcano plot for COG between L. booriae and L. welshimeri

volcano_df = pd.DataFrame(columns=['log2 (fold change)', '-log10 (adjusted P)'])
i =0
for key, val in fdr_cog.items():
    # Merge COG data with group data
    prep_gp = pd.merge(df_cog_prop_t[[key]], group, left_index=True, right_index=True)

    # Calculate median values for positive and negative groups
    median_boo = prep_gp[prep_gp['group'] == 'L. booriae'][key].median()
    median_wels = prep_gp[prep_gp['group'] == 'L. welshimeri'][key].median()

    # Calculate median difference
    median_diff = np.log2((median_boo + 1e-10) / (median_wels + 1e-10))

    # Calculate -log10(p-value)
    neg_log_pval = -np.log10(val)

    # Append the results to the volcano_df DataFrame
    volcano_df.loc[key] = [median_diff, neg_log_pval]
```

```python
# Selecting significant COG based on p-value threshold

pval_threshold = 0.05
log_threshold = -np.log10(pval_threshold)
COG_boo_wels = volcano_df[volcano_df['-log10 (adjusted P)'] > log_threshold]['-log10 (adjusted P)'].to_dict()

plot_volcano_with_stats(
    data=volcano_df,
    x='log2 (fold change)',
    y='-log10 (adjusted P)',
    sig_data=COG_boo_wels,
    title="Differences in COG relative abundance\nbetween L. welshimeri and L. booriae",
    figsize=(6.5, 5.5),
    top_n=21,
    xlim=(-1.5, 1.5),
    ylim=(-1, 40),
    save_as="2a_gen_volcano_cog.pdf",
    out_path=out_path_2,
    point_size=200
)
```

```python
def data_cleaner(count_data, sep):
  def extract_new_name(col_name, sep, unknown_counter=None):
    # Splitting the string at 'p__'
    if col_name == "index":
        return col_name
    parts = col_name.split(sep)
    if len(parts) > 1:
        # Further splitting at ';' and taking the first part
        new_name = parts[1].split(';')[0]
        if len(new_name) <= 1:
            new_name = 'Unknown'
    else:
        new_name = 'Unknown'
    # Ensuring uniqueness for 'Unknown' columns
    if new_name == 'Unknown' and unknown_counter is not None:
        new_name += f"_{unknown_counter[0]}"
        unknown_counter[0] += 1
    return new_name

  # Initialize a counter for 'Unknown' names
  unknown_counter = [1]

  # Renaming the columns with a unique identifier for 'Unknown'
  new_column_names = {col: extract_new_name(col, sep, unknown_counter=unknown_counter) for col in count_data.columns}

  count_data_cleaned = count_data.rename(columns=new_column_names)


  # Identify all 'Unknown' columns
  unknown_columns = [col for col in count_data_cleaned.columns if 'Unknown' in col]


  # sum unknown columns into a new column
  if unknown_columns:
      count_data_cleaned['Unknown'] = count_data_cleaned[unknown_columns].sum(axis=1)

  count_data_cleaned = count_data_cleaned.drop(columns=unknown_columns)

  # Calculate the sum of each column
  row_sums = count_data_cleaned.iloc[:, 1:].sum(axis=1)

  # Divide each entry of each column by the respective row sum
  count_data_cleaned.iloc[:, 1:] = count_data_cleaned.iloc[:, 1:].div(row_sums, axis=0)

  count_data_cleaned.drop('Unknown', axis=1, inplace=True, errors='ignore')

  return count_data_cleaned
```

```python
# Function for making ellipse
def plot_point_cov(points, nstd=2, ax=None, **kwargs):
    """
    Plots an `nstd` sigma ellipse based on the mean and covariance of a point
    "cloud" (points, an Nx2 array).

    Parameters
    ----------
        points : An Nx2 array of the data points.
        nstd : The radius of the ellipse in numbers of standard deviations.
            Defaults to 2 standard deviations.
        ax : The axis that the ellipse will be plotted on. Defaults to the
            current axis.
        Additional keyword arguments are pass on to the ellipse patch.

    Returns
    -------
        A matplotlib ellipse artist
    """
    pos = points.mean(axis=0)
    cov = np.cov(points, rowvar=False)
    return plot_cov_ellipse(cov, pos, nstd, ax, **kwargs)

def plot_cov_ellipse(cov, pos, nstd=2, ax=None, **kwargs):
    """
    Plots an `nstd` sigma error ellipse based on the specified covariance
    matrix (`cov`). Additional keyword arguments are passed on to the
    ellipse patch artist.

    Parameters
    ----------
        cov : The 2x2 covariance matrix to base the ellipse on
        pos : The location of the center of the ellipse. Expects a 2-element
            sequence of [x0, y0].
        nstd : The radius of the ellipse in numbers of standard deviations.
            Defaults to 2 standard deviations.
        ax : The axis that the ellipse will be plotted on. Defaults to the
            current axis.
        Additional keyword arguments are pass on to the ellipse patch.

    Returns
    -------
        A matplotlib ellipse artist
    """
    def eigsorted(cov):
        vals, vecs = np.linalg.eigh(cov)
        order = vals.argsort()[::-1]
        return vals[order], vecs[:,order]

    if ax is None:
        ax = plt.gca()

    vals, vecs = eigsorted(cov)
    theta = np.degrees(np.arctan2(*vecs[:,0][::-1]))

    # Width and height are "full" widths, not radius
    width, height = 2 * nstd * np.sqrt(vals)
    ellip = Ellipse(xy=pos, width=width, height=height, angle=theta, linewidth=2.0, **kwargs)

    ax.add_artist(ellip)
    return ellip
```

#### Using interative variation inflation factor (VIF) to assess multicolinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calculate_vif(df, threshold=10.0, verbose=True):
    df_unique = df.drop_duplicates()
    df_unique = df_unique.select_dtypes(include=[np.number])
    
    variables = df_unique.columns.tolist()
    removed_vars = []
    dropped = True
    
    while dropped and len(variables) > 0:
        dropped = False
        vif_data = pd.DataFrame()
        vif_data["Variable"] = variables
        vif_data["VIF"] = [variance_inflation_factor(df_unique[variables].values, i) 
                           for i in range(len(variables))]
        
        max_vif = vif_data["VIF"].max()
        if max_vif > threshold:
            drop_var = vif_data.loc[vif_data["VIF"].idxmax(), "Variable"]
            if verbose:
                print(f"Dropping '{drop_var}' with VIF={max_vif:.2f}")
            variables.remove(drop_var)
            removed_vars.append(drop_var)
            dropped = True
    
    vif_df = pd.DataFrame()
    vif_df["Variable"] = variables
    vif_df["VIF"] = [variance_inflation_factor(df_unique[variables].values, i) 
                     for i in range(len(variables))]
    
    df_reduced = df_unique[variables].copy()
    
    return df_reduced, vif_df, removed_vars
```

```python
%%R

library(vegan)

out_path_vpa <- "/projects/leaph/yingxian/boo_wels/out/fig_3/vpa_abio_wels_vif_10/"

# Compute adjusted R values
VarPart_val <- function(env_ls, shannon_ls){
  adjr2_ls <- list()

  for (shannon in names(shannon_ls)){
    response <- varpart(
      shannon_ls[[shannon]],
      env_ls$Soil,
      env_ls$Climate,
      env_ls$Landuse
    )

    adjr2_ls[[shannon]] <-
      response$part$indfract$Adj.R.square[
        length(response$part$indfract$Adj.R.square)
      ]
  }

  as.data.frame(adjr2_ls)
}

# Generate VPA plots
VarPart_plot <- function(env_ls, shannon_ls){
  for (shannon in names(shannon_ls)){

    response <- varpart(
      shannon_ls[[shannon]],
      env_ls$Soil,
      env_ls$Climate,
      env_ls$Landuse
    )

    pdf(
      paste0(out_path_vpa, shannon, "_cog_env_vpa.pdf"),
      width = 5,
      height = 5
    )

    plot(
      response,
      digits = 2,
      Xnames = c("Soil", "Climate", "Landuse"),
      bg = c("grey"),
      cex = 1.25
    )

    title(main = shannon)
    dev.off()
  }
}

env_ls <- list(
  Soil     = df_env_sub_wels[, 1:12],
  Climate  = df_env_sub_wels[, 13, drop = FALSE],
  Landuse  = df_env_sub_wels[, 14:21]
)

shannon_ls <- as.list(df_shannon_wels)

# Run VPA
observed_df <- VarPart_val(env_ls, shannon_ls)

observed_r2 <- data.frame(
  COG = colnames(df_shannon_wels),
  Observed_R2 = as.numeric(unlist(observed_df)),
  Explained_var = 1 - as.numeric(unlist(observed_df))
)

write.csv(
  observed_r2,
  paste0(out_path_vpa, "wels_cog_env_residuals.csv"),
  row.names = FALSE
)

# Generate plots
VarPart_plot(env_ls, shannon_ls)
```

```python
from scipy.stats import spearmanr

# Compute Spearman correlation and adjust p-values
def compute_spearman_corr_cog(env_df, richness_df, abiotic_columns, richness_columns):
    corr_results = []
    
    for abiotic_col in abiotic_columns:
        for richness_col in richness_columns:
            if richness_col != "Isolate_ID":
                corr, pval = spearmanr(env_df[abiotic_col], richness_df[richness_col])
                corr_results.append([abiotic_col, richness_col, corr, pval])
    
    # Create a DataFrame
    corr_df = pd.DataFrame(corr_results, columns=["Abiotic", "Richness", "spearmanr_coefficient", "pvalue"])
    
    # Apply FDR correction within each COG richness group
    corr_df["fdr_adjusted_pvalue"] = (
        corr_df.groupby("Richness")["pvalue"]
               .transform(lambda p: multipletests(p, method="fdr_bh")[1])
    )
    
    # Assign significance levels based on adjusted p-values
    conditions = [
        corr_df["fdr_adjusted_pvalue"] < 0.001,
        corr_df["fdr_adjusted_pvalue"] < 0.01,
        corr_df["fdr_adjusted_pvalue"] < 0.05,
        corr_df["fdr_adjusted_pvalue"] >= 0.05,
    ]
    choices = ["***", "**", "*", " "]
    corr_df["asterisk"] = np.select(conditions, choices, default="ns")
    
    return corr_df

# Create pivot tables for heatmap
def prepare_pivot_tables_rich(corr_df):
    # Spearman correlation coefficients
    pivot_corr = corr_df.pivot(index="Abiotic", columns="Richness", values="spearmanr_coefficient")
    
    # Significance asterisks
    pivot_asterisks = corr_df.pivot(index="Abiotic", columns="Richness", values="asterisk")
    
    return pivot_corr, pivot_asterisks
```

```python
from scipy.stats import spearmanr
from statsmodels.stats.multitest import multipletests
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Compute Spearman correlation and adjust p-values
def compute_spearman_correlation(env_df, richness_df, abiotic_columns, richness_columns, group_label):
    corr_results = []
    for abiotic_col in abiotic_columns:
        for richness_col in richness_columns:
            if richness_col != "Isolate_ID":
                corr, pval = spearmanr(env_df[abiotic_col], richness_df[richness_col])
                corr_results.append([abiotic_col, group_label, corr, pval])
    
    corr_df = pd.DataFrame(corr_results, columns=["Abiotic", "Group", "spearmanr_coefficient", "raw_pvalue"])
    
    # Adjust p-values using FDR
    _, fdr_adjusted_pvalues, _, _ = multipletests(corr_df["raw_pvalue"], method="fdr_bh")
    corr_df["fdr_adjusted_pvalue"] = fdr_adjusted_pvalues
    
    # Add asterisk significance labels
    conditions = [
        corr_df["fdr_adjusted_pvalue"] < 0.001,
        corr_df["fdr_adjusted_pvalue"] < 0.01,
        corr_df["fdr_adjusted_pvalue"] < 0.05,
        corr_df["fdr_adjusted_pvalue"] >= 0.05,
    ]
    choices = ["***", "**", "*", " "]
    corr_df["asterisk"] = np.select(conditions, choices, default="ns")
    
    return corr_df

# Create pivot tables for heatmap
def prepare_pivot_tables_grouped(corr_df):
    pivot_corr = corr_df.pivot(index="Abiotic", columns="Group", values="spearmanr_coefficient")
    pivot_asterisks = corr_df.pivot(index="Abiotic", columns="Group", values="asterisk")
    return pivot_corr, pivot_asterisks

# Plot heatmap
def plot_heatmap(pivot_corr, pivot_asterisks, title, output_path=None, cmap="PuOr", vmin=-1, vmax=1, figsize=(14, 10)):
    """
    Plot a heatmap for the Spearman correlation results with significance annotations.

    Parameters:
    - pivot_corr: DataFrame containing the Spearman correlation coefficients.
    - pivot_asterisks: DataFrame containing significance level annotations.
    - title: Title of the heatmap.
    - output_path: Path to save the heatmap. If None, the plot is not saved.
    - cmap: Colormap for the heatmap.
    - vmin: Minimum value for the color scale.
    - vmax: Maximum value for the color scale.
    - figsize: Tuple specifying the figure size (width, height).
    """
    plt.figure(figsize=figsize)
    ax = sns.heatmap(
        pivot_corr,
        annot=pivot_asterisks,
        annot_kws={"size": 10},
        fmt="",
        cmap=cmap,
        center=0,
        cbar_kws={"label": "Spearman coefficient", "pad": 0.02, "shrink": 1, "aspect": 40},  # Adjust aspect for the colorbar
        vmin=vmin,
        vmax=vmax,
    )
    cbar = ax.collections[0].colorbar
    cbar.ax.yaxis.label.set_size(16)
    cbar.ax.tick_params(labelsize=14)
    
    plt.title(title, fontsize=18)
    plt.xticks(rotation=0, ha="center", fontsize=14)
    plt.yticks(fontsize=14)
    plt.xlabel("")
    plt.ylabel("")
    plt.tight_layout()
    
    if output_path:
        plt.savefig(output_path, bbox_inches="tight", dpi=300)
    plt.show()
```

```python
def data_cleaner(count_data, sep):
  def extract_new_name(col_name, sep, unknown_counter=None):
    # Splitting the string at 'p__'
    if col_name == "index":
        return col_name
    parts = col_name.split(sep)
    if len(parts) > 1:
        # Further splitting at ';' and taking the first part
        new_name = parts[1].split(';')[0]
        if len(new_name) <= 1:
            new_name = 'Unknown'
    else:
        new_name = 'Unknown'
    # Ensuring uniqueness for 'Unknown' columns
    if new_name == 'Unknown' and unknown_counter is not None:
        new_name += f"_{unknown_counter[0]}"
        unknown_counter[0] += 1
    return new_name

  # Initialize a counter for 'Unknown' names
  unknown_counter = [1]

  # Renaming the columns with a unique identifier for 'Unknown'
  new_column_names = {col: extract_new_name(col, sep, unknown_counter=unknown_counter) for col in count_data.columns}

  count_data_cleaned = count_data.rename(columns=new_column_names)


  # Identify all 'Unknown' columns
  unknown_columns = [col for col in count_data_cleaned.columns if 'Unknown' in col]


  # sum unknown columns into a new column
  if unknown_columns:
      count_data_cleaned['Unknown'] = count_data_cleaned[unknown_columns].sum(axis=1)

  count_data_cleaned = count_data_cleaned.drop(columns=unknown_columns)

  # Calculate the sum of each column
  row_sums = count_data_cleaned.iloc[:, 1:].sum(axis=1)

  # Divide each entry of each column by the respective row sum
  count_data_cleaned.iloc[:, 1:] = count_data_cleaned.iloc[:, 1:].div(row_sums, axis=0)

  count_data_cleaned.drop('Unknown', axis=1, inplace=True, errors='ignore')

  return count_data_cleaned
```

```python
from scipy.stats import linregress

def plot_distance_decay(
    data,
    group=None,
    type_=None,
    x='dis',
    y='ANIB',
    y_label='',
    point_color="#5DADE2",
    point_size=20,
    title=None,
    out_path='decay_plot.pdf',
    ax=None,
    ylim=None
):
    # --- Filter data if group/type specified ---
    df_plot = data.copy()
    if group is not None:
        df_plot = df_plot[df_plot['group'] == group]
    if type_ is not None:
        df_plot = df_plot[df_plot['type'] == type_]

    if df_plot.empty:
        print(f"No data for group '{group}' and type '{type_}'.")
        return

    # --- Linear regression ---
    slope, intercept, r_value, p_value, std_err = linregress(df_plot[x], df_plot[y])
    r_squared = r_value ** 2

    # --- Plot setup ---
    if ax is None:
        plt.figure(figsize=(4.5, 4))
        ax = plt.gca()

    # --- Scatter plot ---
    sns.scatterplot(
        data=df_plot,
        x=x,
        y=y,
        color=point_color,
        s=point_size,
        edgecolor=None,
        linewidth=0,
        ax=ax
    )

    # --- Regression line and 95% CI ---
    x_sorted = np.sort(df_plot[x])
    y_pred = slope * x_sorted + intercept

    n = len(df_plot[x])
    y_residuals = df_plot[y] - (slope * df_plot[x] + intercept)
    residual_std = np.std(y_residuals, ddof=1)
    sem = residual_std / np.sqrt(n)
    ci = 1.96 * sem  # 95% CI

    # Regression line
    ax.plot(x_sorted, y_pred, color="gray", alpha=0.7, linewidth=2)
    # Shaded 95% CI
    ax.fill_between(x_sorted, y_pred - ci, y_pred + ci, color='gray', alpha=0.2)

    # --- Display slope and R on plot ---
    slope_label = f"Slope = {slope:.3e}"
    r2_label = f"$R^2$ = {r_squared:.2f}"
    ax.text(
        0.98, 0.98,
        slope_label + '\n' + r2_label,
        ha='right',
        va='top',
        transform=ax.transAxes,
        fontsize=10,
        color='black'
    )

    # --- Labels and aesthetics ---
    ax.set_xlabel("Geographic distance (km)", fontsize=12)
    ax.set_ylabel(y_label, fontsize=12)
    plot_title = title if title is not None else f"{group} | {type_}"
    ax.set_title(plot_title, fontsize=13)

    ax.tick_params(axis='both', labelsize=11)
    ax.grid(False)

    if ylim is not None:
        ax.set_ylim(ylim)

    # --- Save plot (only if standalone) ---
    if out_path is not None and ax is plt.gca():
        plt.tight_layout()
        plt.savefig(out_path, format='pdf', dpi=300)
        plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from statannotations.Annotator import Annotator
from scipy.stats import mannwhitneyu, kruskal
from statsmodels.stats.multitest import multipletests

def plot_3_box_with_mw_and_swarm(
    data, x, y, box_pairs, title="", figsize=(4, 6), 
    colors=['#AF7AC5', '#D08974', '#E1712B'], 
    boxplot_kwargs=None, out_path="", xlim=None
):
    """
    Plots a box + swarm plot with FDR-corrected p-values annotated.
    Also prints the Kruskal-Wallis test p-value.

    Parameters:
    - data (DataFrame): The data to plot.
    - x (str): Column for group/category (x-axis).
    - y (str): Column for numeric values (y-axis).
    - box_pairs (list of tuple): Pairs of groups to compare.
    - title (str): Title of the plot.
    - figsize (tuple): Figure size in inches.
    - colors (list): Colors for the box and swarm plots.
    - boxplot_kwargs (dict): Custom kwargs for the boxplot.
    - out_path (str): Path to save the output figure.
    - xlim (tuple): Optional x-axis limits.
    """

    # Default boxplot styling
    if boxplot_kwargs is None:
        boxplot_kwargs = {
            'boxprops':     {'edgecolor': 'k', 'linewidth': 2.5, 'facecolor': 'white'},
            'medianprops':  {'color': 'k', 'linewidth': 2.5},
            'whiskerprops': {'color': 'k', 'linewidth': 2.5},
            'capprops':     {'color': 'k', 'linewidth': 2.5},
            'width': 0.5
        }

    # Define group order
    boxplot_order = ["Within soil", "Soil-wildlife", "Within wildlife"]

    # Remove rows with NaN y-values
    data = data.dropna(subset=[y])
    if data.empty:
        print(f"No data to plot for {title} after filtering out NaN rows.")
        return

    # Compute and print Kruskal-Wallis p-value
    grouped_data = [data[data[x] == group][y] for group in boxplot_order]
    kw_stat, kw_p = kruskal(*grouped_data)
    print(f"Kruskal-Wallis test p-value for {title}: {kw_p:.4g}")

    # Prepare figure
    plt.rcParams["figure.figsize"] = figsize
    fig, ax = plt.subplots(figsize=figsize)

    # Plot boxplot
    sns.boxplot(data=data, x=x, y=y,
                order=boxplot_order,
                palette=dict(zip(boxplot_order, colors)),
                ax=ax,
                **boxplot_kwargs, showfliers=False)

    # Plot swarm
    sns.swarmplot(data=data, x=x, y=y,
                  order=boxplot_order,
                  palette=dict(zip(boxplot_order, colors)),
                  ax=ax,
                  size=7, edgecolor='k', linewidth=0.5)

    # Compute raw p-values (Mann-Whitney U)
    raw_pvals = []
    for g1, g2 in box_pairs:
        vals1 = data[data[x] == g1][y]
        vals2 = data[data[x] == g2][y]
        _, p = mannwhitneyu(vals1, vals2, alternative='two-sided')
        raw_pvals.append(p)

    # FDR correction
    _, qvals, _, _ = multipletests(raw_pvals, method='fdr_bh')

    # Annotate with corrected p-values
    annotator = Annotator(ax, box_pairs, data=data, x=x, y=y, order=boxplot_order)
    annotator.configure(text_format='simple', loc='outside', pvalue_format_string="{:.2g}")
    annotator.set_pvalues_and_annotate(qvals)

    # Apply x-axis limits if provided
    if xlim is not None:
        ax.set_xlim(xlim)

    # Labels and style
    plt.xticks(rotation=50, size=12)
    plt.yticks(rotation=0, size=12)
    plt.ylabel('%', size=15)
    plt.xlabel(' ', size=15)
    plt.tight_layout()

    # Save output
    safe_title = title.replace('/', '_').replace('\\', '_')
    if out_path:
        plt.savefig(out_path + f'{safe_title}.pdf', format='pdf', dpi=600)

    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.multitest import multipletests
from statannotations.Annotator import Annotator

def plot_2_box_with_swarm(
    data,
    x,
    y,
    title="",
    figsize=(3, 5),
    swarm_size=3,
    colors=["#EC7063", "#5499C7"],
    outpath=None,
    xlabel="",
    ylabel="",
    remove_outliers=True,
    ax=None
):
    """
    Plots two combined box plots with swarm plots and performs statistical annotation with FDR BH correction.
    
    Parameters:
    - data (DataFrame): The data to plot.
    - x (str): The name of the column in 'data' to be used as x-axis groups.
    - y (str): The name of the column in 'data' for the y-axis values.
    - title (str): Title of the plot.
    - figsize (tuple): Size of the figure (ignored if ax is provided).
    - colors (list): List of colors for the box plots.
    - outpath (str): The path where the plot should be saved.
    - xlabel (str): Label for the x-axis.
    - ylabel (str): Label for the y-axis.
    - remove_outliers (bool): Whether to remove outliers based on IQR.
    - ax (matplotlib.axes.Axes): Optional matplotlib Axes to plot into.
    """

    # Remove outliers using IQR method if required
    if remove_outliers:
        Q1 = data[y].quantile(0.25)
        Q3 = data[y].quantile(0.75)
        IQR = Q3 - Q1
        data = data[(data[y] >= (Q1 - 1.5 * IQR)) & (data[y] <= (Q3 + 1.5 * IQR))]

    # Create figure and axis if not provided
    if ax is None:
        fig, ax = plt.subplots(figsize=figsize)

    sns.boxplot(
        data=data,
        x=x,
        y=y,
        palette=colors,
        ax=ax,
        showfliers=False,
        boxprops=dict(facecolor='none', edgecolor='#000000'),
        whiskerprops=dict(color='#000000'),
        capprops=dict(color='#000000'),
        medianprops=dict(color='#000000'),
    )

    sns.swarmplot(
        data=data,
        x=x,
        y=y,
        size=swarm_size,
        palette=colors,
        ax=ax,
        edgecolor='k', linewidth=0.5
    )

    # Define box pairs for comparison
    unique_groups = data[x].dropna().unique()
    if len(unique_groups) >= 2:
        box_pairs = [(unique_groups[0], unique_groups[1])]

        annotator = Annotator(ax, box_pairs, data=data, x=x, y=y)
        annotator.configure(
            test='Mann-Whitney',
            text_format='full',
            loc='inside',
            verbose=2,
            pvalue_format_string="{:.2e}",
        ).apply_test().annotate()

    # Customize the plot
    ax.set_title(title, fontsize=14)
    ax.set_xlabel(xlabel if xlabel else "", fontsize=12)
    ax.set_ylabel(ylabel if ylabel else y, fontsize=12)
    ax.tick_params(axis='x', labelsize=11)

    # Save the plot if no external axis is passed and outpath is provided
    if ax is None:
        plt.tight_layout()
        if outpath:
            plt.savefig(outpath, format='pdf', dpi=300)
        plt.show()
```
